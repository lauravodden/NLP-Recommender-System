```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
pd.set_option('precision', 0)

resource_catalog = pd.read_csv("resource_catalog_with_FOE.csv", dtype=object)

# Create new empty column
resource_catalog['CourseId'] = ""
#combine words from course name and title
resource_catalog['CourseId'] = resource_catalog.CourseId.str.cat(" "+resource_catalog.course_name.str.cat(" "+resource_catalog.title))

combined_title_course = resource_catalog['CourseId']
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(combined_title_course)
```

```python
# true_k, derived from trial and error
true_k = 350
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=500, n_init=15, random_state=100)
model.fit(X)
```

```python
# Top terms in each cluster.
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :15]:
        print(' %s' % terms[ind]),
    print
```

```python
import pickle
# Save machine learning model
filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
def cluster_predict(str_input):
    Y = vectorizer.transform(list(str_input))
    prediction = model.predict(Y)
    return prediction
```

```python
# create new column in dataframe which is combination of (CourseId, CourseTitle, Description) in existing data-frame
resource_catalog['InputString'] = resource_catalog['title']
# Create new column for storing predicted categories from our trained model.
resource_catalog['ClusterPrediction'] = ""
# Cluster category for each live course
resource_catalog['ClusterPrediction']=resource_catalog.apply(lambda x: cluster_predict(resource_catalog['InputString']), axis=0)
```

```python
def recommend_util(str_input):
    # match on the basis of courseid
    temp_df = resource_catalog.loc[resource_catalog['title'] == str_input]
    
    # Predict category of input string category
    prediction_inp = temp_df.iloc[0][9]

    # Based on prediction 10 random courses are recommended from resource_catalog
    temp_df = resource_catalog.loc[resource_catalog['ClusterPrediction'] == prediction_inp]

    temp_df = temp_df.drop_duplicates(subset ="course_name",keep = "first", inplace = False) 

    rec_courses = temp_df['course_name']
    return rec_courses[1:11]
```

```python
str_input = "feminist cultural studies"
res = recommend_util(str_input)

length = len("Suggested courses for:  ") + len(str_input)

print("Suggested courses for: ", str_input)
i = 0
while i < length:
    print('-', end="")
    i = i + 1
print("\n")    
print(res)

    
```
